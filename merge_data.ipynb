```python
import pandas as pd
```

```python
data = pd.read_csv('data_cleaned_final_1.csv')         ## this is the cleaned csv
data = data.drop('Unnamed: 0', axis=1) 
data.head()
```

```python
reviews_300 = pd.read_csv('reviews1000-1250clean.csv')       ## select the right batch to analyze 
reviews_300 = reviews_300.drop('Unnamed: 0', axis=1)
reviews_300['date'] = [pd.to_datetime(reviews_300.Date[x]) for x in range(len(reviews_300.Date))]
reviews_300 = reviews_300.drop('Date', axis=1)
print(reviews_300.shape)
reviews_300.head()
```

For each year add the precise date of the Grammy ceremony

```python
def grammy(year):
    if year == 2000: day_of_grammy = pd.to_datetime('2000-02-23')
    elif year == 2001: day_of_grammy = pd.to_datetime('2001-02-21')
    elif year == 2002: day_of_grammy = pd.to_datetime('2002-02-27')
    elif year == 2003: day_of_grammy = pd.to_datetime('2003-02-23')
    elif year == 2004: day_of_grammy = pd.to_datetime('2004-02-08')
    elif year == 2005: day_of_grammy = pd.to_datetime('2005-02-13')
    elif year == 2006: day_of_grammy = pd.to_datetime('2006-02-08')
    elif year == 2007: day_of_grammy = pd.to_datetime('2007-02-11')
    elif year == 2008: day_of_grammy = pd.to_datetime('2008-02-10')
    elif year == 2009: day_of_grammy = pd.to_datetime('2009-02-08')
    elif year == 2010: day_of_grammy = pd.to_datetime('2010-01-31')
    elif year == 2011: day_of_grammy = pd.to_datetime('2011-02-13')
    elif year == 2012: day_of_grammy = pd.to_datetime('2012-02-12')
    elif year == 2013: day_of_grammy = pd.to_datetime('2013-02-10')
    elif year == 2014: day_of_grammy = pd.to_datetime('2014-01-26')
    elif year == 2015: day_of_grammy = pd.to_datetime('2015-02-08')
    elif year == 2016: day_of_grammy = pd.to_datetime('2016-02-15')
    elif year == 2017: day_of_grammy = pd.to_datetime('2017-02-12')
    elif year == 2018: day_of_grammy = pd.to_datetime('2018-01-28')
    elif year == 2019: day_of_grammy = pd.to_datetime('2019-02-10')

    return day_of_grammy

data['grammy_date'] = [grammy(data.year[x]) for x in range(len(data.year))]
data.head()
```

URL is used as key to merge the two datasets

```python
import re

def clean(s):
    s = re.sub(r"^\s|\s$", "", s) # replace spaces at the start and at the end of the string
    s = re.sub(r"\s","-",s) # replace normal spaces with dashes
    s = re.sub(r"\.","",s) #remove dots
    s = re.sub(r"&","and",s) # replace & with "and"
    s = re.sub(r"[áàâäāãåăąÀÁÂÄĀÃÅĂĄ]","a",s)
    s = re.sub(r"[éèêëēėęĚĒÈÉÊËĖĘ]","e",s)
    s = re.sub(r"[íìîïīįīĨÌÍÎÏĪĮ]","i",s)
    s = re.sub(r"[óòôöōõøőÓÒÔÖŌÕØŐ]","o",s)
    s = re.sub(r"[úùûüūųÚÙÛÜŪŲ]","u",s)
    s = re.sub(r"[çĉćċčÇĈĆĊČ]","c",s)
    s = re.sub(r"[ñńňņŉŋÑŃŇŅŉŊ]","n",s)
    s = s.lower() #lower case 
    if s == "ti":
        s = "t_i"
    elif s == "nati-cano's-mariachi-los-camperos":
        s = "mariachi-los-camperos-de-nati-cano"
    elif s == "george-carlin-it's-bad-for-ya!":
        s = "george-carlin"
    elif s == "ferdinand-'jelly-roll'-morton":
        s = "jelly-roll-morton"
    elif s == "":
        s = ""
    s = re.sub(r"['!:]","",s) # remove extra punctuation

    if s == "kalani-pea":
        s = "kalani-pe'a"
    
    
    return s
```

```python
def clean_total(l):
    new_artist = []

    # FIRST ROUND OF CLEANING
    for i in l:
        if re.search(r".*\(.*artist.*\).*",str(i)):
            name = " ".join((re.findall(r".*(?=\s\(.*artist.*\))",str(i))))
            if re.search(r"\),",name):
                #print("hello")
                name = " ".join(re.findall(r"(?<=,)[^,]+$",name)) # take everything after the comma
                name = clean(name)
                new_artist.append(name)
            else:
                name = clean(name)
                new_artist.append(name)

        elif re.search(r"\(",i):
            name = " ".join(re.findall(r"^.*?(?<=\()",i))
            name = re.sub(r"\(","",name)
            name = clean(name)
            new_artist.append(name)
        else:
            name = clean(i)
            new_artist.append(name)
    return new_artist
```

```python
data["workers_clean"] = clean_total(data.workers)
data["nominee_semiclean"] = data.nominee.astype(str).apply(lambda x: clean(x)).tolist()
data["url_final"] = ["https://rateyourmusic.com/release/album/" + data.workers_clean[i] + "/" + data.nominee_semiclean[i] + "/" if data.url[i] == '0' else data.url[i] for i in range(len(data))]
data = data.drop('url', axis = 1)
data.head()
```

```python
reviews_300['url_final'] = ["https://rateyourmusic.com/release/album/" + reviews_300['Album'][x] for x in range(len(reviews_300['Album']))]
reviews_300.head()
```

```python
data_merged = data.merge(reviews_300, on = 'url_final', how = 'inner')
print(data_merged.shape)
data_merged.head()
```

Add a dummy --> 0 if review is written before grammy, 1 if it writter after 

```python
data_merged['after_grammy'] = [True if data_merged.date[x] > data_merged.grammy_date[x] else False for x in range(len(data_merged.date))]
data_merged.head()
```

Compute average rating

```python
rating_means = data_merged.groupby(['url_final', 'after_grammy'], as_index=False)['Rating'].mean()
rating_means
```

```python
data_w_ratings = data.merge(rating_means, on = 'url_final')
data_w_ratings.head()
```

```python
data_w_ratings['rating_after'] = data_w_ratings[data_w_ratings.after_grammy]['Rating']
data_w_ratings['rating_before'] = data_w_ratings[~data_w_ratings.after_grammy]['Rating']
```

```python
clean_ratings = data_w_ratings.groupby('url_final').agg({'rating_before': 'first', 'rating_after': 'last'}).reset_index()
clean_ratings
```

Compute average positiveness

```python
sent_means = data_merged.groupby(['url_final', 'after_grammy'], as_index=False)['sentiment_score'].mean()
sent_means
```

```python
data_w_sentiment = data.merge(sent_means, on = 'url_final')
data_w_sentiment.head()
```

```python
data_w_sentiment['sentiment_after'] = data_w_sentiment[data_w_sentiment.after_grammy]['sentiment_score']
data_w_sentiment['sentiment_before'] = data_w_sentiment[~data_w_sentiment.after_grammy]['sentiment_score']
```

```python
clean_sentiment = data_w_sentiment.groupby('url_final').agg({'sentiment_before': 'first', 'sentiment_after': 'last'}).reset_index()
clean_sentiment
```

Compute number of reviews

```python
num_reviews = data_merged.groupby(['url_final', 'after_grammy'], as_index=False)['Review ID'].count().rename(columns = {'Review ID': 'review_num'})
num_reviews
```

```python
data_w_num_reviews = data.merge(num_reviews, on = 'url_final')
data_w_num_reviews.head()
```

```python
data_w_num_reviews['num_reviews_after'] = data_w_num_reviews[data_w_num_reviews.after_grammy]['review_num']
data_w_num_reviews['num_reviews_before'] = data_w_num_reviews[~data_w_num_reviews.after_grammy]['review_num']
```

```python
clean_num_reviews = data_w_num_reviews.groupby('url_final').agg({'num_reviews_before': 'first', 'num_reviews_after': 'last'}).reset_index()
clean_num_reviews
```

Final dataset (for the batch)

```python
data_merged_final = data.merge(clean_ratings, on = 'url_final').merge(clean_sentiment, on = 'url_final').merge(clean_num_reviews, on = 'url_final')
data_merged_final.head()
```
